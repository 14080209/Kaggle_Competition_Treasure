## 自定义评分损失函数

```python
def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred))/ 2.0
    diff        = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return np.mean(diff),diff
```

### 特征提取

```python
def get_features(df, is_test = False): 
    df_fea        = pd.DataFrame() 
    fea_columns   = [column for column in df.columns if column != 'Page']
     
    if is_test == False:
        df_raw    = df[fea_columns].fillna(0).values[:, :-64]
        df_label  = df[fea_columns].fillna(0).values[:, -62:]
    else:
        df_raw    = df[fea_columns].fillna(0).values
    
    day7_median   = []
    day28_median  = []
    day49_median  = []
    day365_median = []
     
    df_fea['Page'] = df['Page'].values
    
    
    len_ = df_raw.shape[1] 
    
    for pr in [7,28,49,365]:
        nums = len_ // pr
        for i in range(nums):
            if pr == 7:
                day7_median.append( (len_-(i+1)*7,len_-i*7) )
            elif pr == 28:
                day28_median.append( (len_-(i+1)*28,len_-i*28) )
            elif pr == 49:
                day49_median.append( (len_-(i+1)*49,nums-i*49) )
            else:
                day365_median.append( (len_-(i+1)*365,len_-i*365) )  
    # 中位数 #   
    median_cols = []
    print('get 7 day median')
    for i,(st,en) in enumerate(day7_median): 
        df_fea['median7_{}'.format(i+1)]   = np.nanmedian(df_raw[:,st:en],axis=1)[:,]
        median_cols.append('median7_{}'.format(i+1))
    print('get 28 day median')
    for i,(st,en) in enumerate(day28_median):
        df_fea['median28_{}'.format(i+1)]  = np.nanmedian(df_raw[:,st:en],axis=1)
    print('get 49 day median')
    for i,(st,en) in enumerate(day49_median):
        df_fea['median49_{}'.format(i+1)]  = np.nanmedian(df_raw[:,st:en],axis=1)
        
    print('get 365 day median')
    for i,(st,en) in enumerate(day365_median):
        df_fea['median365_{}'.format(i+1)] = np.nanmedian(df_raw[:,st:en],axis=1)
    
    # 中位数的中位数 #
    df_fea['median_median'] = np.nanmedian(df_fea[median_cols].values,axis=1)
    
    if is_test:
        return df_fea
    else:
        return df_fea,df_label  
```

```python
median7_cols       = [col for col in df_fea.columns if 'median7' in col]
median28_cols      = [col for col in df_fea.columns if 'median28' in col]
median49_cols      = [col for col in df_fea.columns if 'median49' in col]
median365_cols     = [col for col in df_fea.columns if 'median365' in col]
median_median_cols = [col for col in df_fea.columns if 'median_median' in col]
```

```python
from tensorflow.keras.layers import GRU, Dense, Input, Lambda, Reshape, Conv1D, Activation,Dropout
from tensorflow.keras.layers import concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
import tensorflow.keras.backend as K

ESTIMATES_COUNT           = 7
TRUNCATED_SEQUENCE_LENGTH = 149

Median7_LEN       = len(median7_cols)
Median28_LEN      = len(median28_cols)
Median49_LEN      = len(median49_cols)
Median365_LEN     = len(median365_cols)
MedianMedian_LEN  = len(median_median_cols)

PREDICTION_LENGTH = 60
```

```python
def smape_loss(y_true, y_pred):
    diff = K.abs((y_true - y_pred) / K.clip(K.abs(y_true) + K.abs(y_pred), K.epsilon(), None))
    return 200 * K.mean(diff, axis=-1)

def combine_estimates(data):
    estimates, coefs = data
    return K.sum(estimates * coefs, axis=-1)

# feed visit counts during latest TRUNCATED_SEQUENCE_LENGTH days here
median_7_data_raw         = Input(shape=(Median7_LEN,))
median_28_data_raw        = Input(shape=(Median28_LEN,))
median_49_data_raw        = Input(shape=(Median49_LEN,))
median_365_data_raw       = Input(shape=(Median365_LEN,))
median_median_data_raw    = Input(shape=(MedianMedian_LEN,))

median_7_data      = Reshape((Median7_LEN, 1))(median_7_data_raw)
median_28_data     = Reshape((Median28_LEN, 1))(median_28_data_raw)
median_49_data     = Reshape((Median49_LEN, 1))(median_49_data_raw)
median_365_data    = Reshape((Median365_LEN, 1))(median_365_data_raw)
median_median_data = Reshape((MedianMedian_LEN, 1))(median_median_data_raw)

conv_result1 = Activation('relu')(concatenate([Conv1D(32, 3, padding='same')(median_7_data),\
                                              Conv1D(32, 3, dilation_rate=3, padding='same')(median_7_data),]))
conv_result1 = Activation('relu')(concatenate([Conv1D(32, 3, padding='same')(conv_result1),\
                                              Conv1D(32, 3, dilation_rate=3, padding='same')(conv_result1),]))

rnn_result1 = GRU(128)(conv_result1)
res1        = Dense(1)(rnn_result1)

conv_result2 = Activation('relu')(concatenate([Conv1D(32, 3, padding='same')(median_28_data),\
    Conv1D(32, 3, dilation_rate=3, padding='same')(median_28_data),]))
conv_result2 = Activation('relu')(concatenate([Conv1D(32, 3, padding='same')(conv_result2),\
    Conv1D(32, 3, dilation_rate=3, padding='same')(conv_result2),]))

rnn_result2 = GRU(64)(conv_result2)
res2        = Dense(1)(rnn_result2)

conv_result3 = Activation('relu')(Conv1D(8, 3, padding='same')(median_49_data)) 
conv_result3 = (Reshape((conv_result3.shape[1] * conv_result3.shape[2],))(conv_result3))
res3         = Dense(1)(conv_result3)
 

res        = concatenate([Reshape((1,1))(res1), Reshape((1,1))(res2), Reshape((1,1))(res3), median_365_data, median_median_data],axis=1)
res        = (Reshape((res.shape[1] * res.shape[2],))(res)) 
result     = Dense(62)(res)

model = Model(inputs=[median_7_data_raw, median_28_data_raw,median_49_data_raw,median_365_data_raw,median_median_data_raw], outputs= result)
model.compile(optimizer=Adam(lr=1e-3), loss=smape_loss)
```

```python
def get_features_3(df, is_test = False): 
    df_fea        = pd.DataFrame() 
    fea_columns   = [column for column in df.columns if column != 'Page']
     
    if is_test == False:
        df_raw    = df[fea_columns].fillna(0).values[:, :-64]
        df_label  = df[fea_columns].fillna(0).values[:, -62:]
    else:
        df_raw    = df[fea_columns].fillna(0).values
    
    day7_median   = []
    day28_median  = []
    day49_median  = []
    day365_median = []
     
    df_fea['Page'] = df['Page'].values
    
    
    len_ = df_raw.shape[1] 
    
    for pr in [7,28,49,365]:
        nums = len_ // pr
        for i in range(nums):
            if pr == 7:
                day7_median.append( (len_-(i+1)*7,len_-i*7) )
            elif pr == 28:
                day28_median.append( (len_-(i+1)*28,len_-i*28) )
            elif pr == 49:
                day49_median.append( (len_-(i+1)*49,nums-i*49) )
            else:
                day365_median.append( (len_-(i+1)*365,len_-i*365) )  
    # 中位数 #   
    median_cols = []
    
    for i,j in enumerate(range(100)): 
        df_fea['last_{}'.format(i+1)]   =  df_raw[:,-j-1]
    
    print('get 7 day median')
    for i,(st,en) in enumerate(day7_median): 
        df_fea['median7_{}'.format(i+1)]   = np.nanmedian(df_raw[:,st:en],axis=1)[:,] 
        median_cols.append('median7_{}'.format(i+1))
    print('get 28 day median')
    for i,(st,en) in enumerate(day28_median):
        df_fea['median28_{}'.format(i+1)]  = np.nanmedian(df_raw[:,st:en],axis=1) 
    print('get 49 day median')
    for i,(st,en) in enumerate(day49_median):
        df_fea['median49_{}'.format(i+1)]  = np.nanmedian(df_raw[:,st:en],axis=1)
         
    print('get 365 day median')
    for i,(st,en) in enumerate(day365_median):
        df_fea['median365_{}'.format(i+1)] = np.nanmedian(df_raw[:,st:en],axis=1)
    
    # 中位数的中位数 #
    df_fea['median_median'] = np.nanmedian(df_fea[median_cols].values,axis=1)
    
    if is_test:
        return df_fea
    else:
        return df_fea,df_label  
```
