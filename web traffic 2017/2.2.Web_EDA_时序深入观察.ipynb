## 深度信息观察
### 同类网站观察
wiki网站的网址有一个比较大的特性,就是前面往往是国家的信息,即(国家.wikipedia.org),我们发现：

- 一共有7种语言再加上其他页面。这里使用的语言有：英语、日语、德语、法语、汉语、俄语和西班牙语。
- 英语的占比是最多的

```python
import re
from collections import Counter
def get_country(page):
    res = re.search('[a-z][a-z].wikipedia.org',page)
    if res:
        return res[0][0:2]
    return 'na'

train['lang'] = train.Page.map(get_country)
```

### 不同国家的wiki网站的流量全局分析 
#### 流量总体之和观察
- 我们发现由于英语每日的浏量比其他语言的要多很多，这应该是维基百科是一个美国网站的原因。
- 此外我们发现英国和俄罗斯的地块在第400天（2016年8月左右）出现了非常大的峰值，2016年晚些时候英国的数据又出现了几个峰值。这可能是8月份夏季奥运会和美国大选的结果。
- 西班牙的数据有一个清晰的周期结构，好像是大约1周的快速周期，每6个月左右有一个明显的下降。
- 中国和Media的流量是最低的

```python
lang_sets = {}
lang_sets['en'] = train[train.lang=='en'].iloc[:,0:-1]
lang_sets['ja'] = train[train.lang=='ja'].iloc[:,0:-1]
lang_sets['de'] = train[train.lang=='de'].iloc[:,0:-1]
lang_sets['na'] = train[train.lang=='na'].iloc[:,0:-1]
lang_sets['fr'] = train[train.lang=='fr'].iloc[:,0:-1]
lang_sets['zh'] = train[train.lang=='zh'].iloc[:,0:-1]
lang_sets['ru'] = train[train.lang=='ru'].iloc[:,0:-1]
lang_sets['es'] = train[train.lang=='es'].iloc[:,0:-1]

mean_flows = {}
for key in lang_sets:
    mean_flows[key] = lang_sets[key].iloc[:,1:].sum(axis=0) / lang_sets[key].shape[0]
```

```python
mean_flows.keys()
```

```python
days = [r for r in range(mean_flows['en'].shape[0])]

fig = plt.figure(1,figsize=[12,10])
plt.ylabel('Views per Page')
plt.xlabel('Day')
plt.title('Pages in Different Languages')
labels={'en':'English','ja':'Japanese','de':'German','na':'Media','fr':'French','zh':'Chinese','ru':'Russian','es':'Spanish'}

for key in mean_flows:
    plt.plot(days,mean_flows[key],label = labels[key])
    
plt.legend()
plt.show()
```

### 流量的周期性观察(采用FFT)
通过上面的观察，我们发现好像每个国家的数据或多或少存在一些周期性，为了能更好的观察这些数据的周期性结构，我们分别画出每一个傅里叶变化的大续爱情况。快速傅立叶变换的峰值显示了周期信号中最强的频率。

- 我们发现，西班牙的周期性是最明显的；
- 其他的一些国家的web流量也存在周期性；
- 俄罗斯的和Media的并没有显示出非常强的周期性；


```python
from scipy.fftpack import fft
def plot_with_fft(key):
    fig = plt.figure(1,figsize=[12,8])
    plt.ylabel('Views per Page')
    plt.xlabel('Day')
    plt.title(labels[key])
    plt.plot(days,mean_flows[key],label = labels[key] )
    
    fig         = plt.figure(2,figsize=[12,8])
    fft_complex = fft(mean_flows[key]) # 傅里叶转换
    fft_mag     = [np.sqrt(np.real(x)*np.real(x)+ np.imag(x)*np.imag(x)) for x in fft_complex]
    fft_xvals   = [day / days[-1] for day in days] 
    
    npts        = len(fft_xvals) // 2 + 1
    
    fft_mag     = fft_mag[:npts]
    fft_xvals   = fft_xvals[:npts]
        
    plt.ylabel('FFT Magnitude')
    plt.xlabel(r"Frequency [days]$^{-1}$")
    plt.title('Fourier Transform')
    plt.plot(fft_xvals[1:],fft_mag[1:],label = labels[key] )
    # Draw lines at 1, 1/2, and 1/3 week periods
    plt.axvline(x=1./7,color='red',alpha=0.3)
    plt.axvline(x=2./7,color='red',alpha=0.3)
    plt.axvline(x=3./7,color='red',alpha=0.3)

    plt.show()

for key in mean_flows:
    plot_with_fft(key)
```

### 代表性的网页观察
#### 流量最多网页观察
此处我们抽样出现次数最多的网页作为代表进行观察，当然大家也可以抽样波动最大的网页作为代表等

```python
# For each language get highest few pages
npages = 5
top_pages = {}
for key in lang_sets:
    print(key)
    sum_set          = pd.DataFrame(lang_sets[key][['Page']])
    sum_set['total'] = lang_sets[key].sum(axis=1)
    sum_set          = sum_set.sort_values('total',ascending=False)
    print(sum_set.head(10))
    top_pages[key] = sum_set.index[0]
    print('\n\n')
```

下面我们在单独看看每个国家总浏览量最多的一个网页，

- 我们发现en的是主页浏览量最多，这是可以理解的，就像很多网站，我们都会先进入主页再搜索刚兴趣的内容一样。
- 俄罗斯的可能受到奥运会的影响，所以有一段时间流量非常大，但是之后趋于平静；
- 西班牙的流量也很奇怪，最大的网页流量不断下跌。 

```python
for key in top_pages:
    fig = plt.figure(1,figsize=(12,8))
    cols = train.columns
    cols = cols[1:-1]
    data = train.loc[top_pages[key],cols]
    plt.plot(days,data)
    plt.xlabel('Days')
    plt.ylabel('Views')
    plt.title(train.loc[top_pages[key],'Page'])
    plt.show()
```

#### PCF & PACF观察
我们发现，
- 在大多数情况下，由于每周效应，每隔一周都会有很强的相关性或者抗相关性；
- 对于部分自相关，第一周一般是最强的，然后慢慢开始缓下来。

```python
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import acf

for key in top_pages:
    fig = plt.figure(1,figsize=[10,5])
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    cols = train.columns[1:-1]
    data = np.array(train.loc[top_pages[key],cols])
    data_diff = [data[i] - data[i-1] for i in range(1,len(data))]
    autocorr  = acf(data_diff)
    pac       = pacf(data_diff)

    x = [x for x in range(len(pac))]
    ax1.plot(x[1:],autocorr[1:])

    ax2.plot(x[1:],pac[1:])
    ax1.set_xlabel('Lag')
    ax1.set_ylabel('Autocorrelation')
    ax1.set_title(train.loc[top_pages[key],'Page'])

    ax2.set_xlabel('Lag')
    ax2.set_ylabel('Partial Autocorrelation')
    plt.show()
```

### Arima实验
- 此处我们通过Arima对我们的数据进行训练观察，我们发现，其实Arima模型在很多网页上预测的结果都看上去还是挺不错的，可以作为一个基本的baseline. 但是网页比较多，而Arima的训练需要比较多的时间，所以需要一定的机器资源。

```python
from statsmodels.tsa.arima_model import ARIMA
import warnings

cols = train.columns[1:-1]
# plt.figure(figsize=[12,8])
for key in top_pages:
    data = np.array(train.loc[top_pages[key],cols],'f')
    result = None
    with warnings.catch_warnings():
        warnings.filterwarnings('ignore')
        try:
            arima  = ARIMA(data,[2,1,4])
            result = arima.fit(disp=False)
        except:
            try:
                arima  = ARIMA(data,[2,1,2])
                result = arima.fit(disp=False)
            except:
                print(train.loc[top_pages[key],'Page'])
                print('\tARIMA failed')
    #print(result.params)
    pred = result.predict(2,599,typ='levels')
    x = [i for i in range(600)]
    i=0

    plt.plot(x[2:len(data)],data[2:] ,label='Data')
    plt.plot(x[2:],pred,label='ARIMA Model')
    plt.title(train.loc[top_pages[key],'Page'])
    plt.xlabel('Days')
    plt.ylabel('Views')
    plt.legend()
    plt.show() 
```
